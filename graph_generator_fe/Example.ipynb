```python
import graph_generator as gg
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

#read in image data (this example uses a CASA-style .fits image which includes multiple polarizations)
#to load in polarization DIFMAP style, just define file paths to stokes_q and stokes_u input parameters
#Example difmap_pol_image=gg.ImageData(fits_file="stokes_i.fits",stokes_q="stokes_q.fits",stokes_u="stokes_u.fits")
pol_image=gg.ImageData(fits_file="0506_final_pol.fits")

#plot total intensity
gg.FitsImage(pol_image,stokes_i_sigma_cut=4)
plt.show()
```

```python
#you can also use a color map if you like and set plotlims
gg.FitsImage(pol_image,
             stokes_i_sigma_cut=3,
             im_colormap=True,
             im_color="inferno",
             contour_color="white",
             xlim=[10,-10],
             ylim=[-12,8]
            )
plt.show()
```

```python
#plot linear polarization
plot=gg.FitsImage(pol_image,
                  plot_mode="lin_pol", #set plot mode lin_pol
                  title="My Polarization Map", #add custom title
                  xlim=[6,-6], #set plot limits
                  ylim=[-8,4], #set plot limits
                  im_color="RdPu" #choose colormap
                 )

plt.show()
```

```python
#plot fractional polarization
plot=gg.FitsImage(pol_image,
                  plot_mode="frac_pol",
                  im_color="RdBu",
                  plot_beam=False, #hide beam
                  evpa_distance=15, #evpa distance in pixel
                  evpa_len=10, #evpa length in pixel
                  lin_pol_sigma_cut=3 #sigma cut for linear polarization to mask
                 )
plot.ax.set_xlim(6,-6)
plot.ax.set_ylim(-8,4)
plt.show()
```

```python
#plot modelfit components

#read in image data
image=gg.ImageData(fits_file="modelfit_final1.fits",model="modelfit_final1.fits")

#plot image with model components
plot=gg.FitsImage(image,overplot_gauss=True)

#change plot limits
plot.ax.set_xlim(6,-6)
plot.ax.set_ylim(-8,4)
plt.show()
```

```python
#plot clean components (example data from MOJAVE database)
mojave_image=gg.ImageData(fits_file="0528+134.u.2023_09_24.icn.fits",model="0528+134.u.2023_09_24.icn.fits")

plot=gg.FitsImage(mojave_image,overplot_clean=True)
plt.show()
```
